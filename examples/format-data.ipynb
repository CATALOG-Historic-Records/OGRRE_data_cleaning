```python
import json
import pandas as pd
from pathlib import Path
```

```python

FILE_NAME = 'OGRRE.json'
WRITE_JSON = False       # Write the reformatted json file (for easier viewing)
WRITE_COLUMNS = True     # Write the unique columns and their counts to json
WRITE_DATAFRAME = True   # Write the dataframe to a csv file

path = Path(FILE_NAME)

# Open the data file reformat json file for easier viewing if desired
with open(path, 'r') as fr:
    data = json.load(fr)

    if WRITE_JSON:
        json.dump(data, open('{}-formatted.json'.format(path.stem), 'w'), indent=4)
```

```python
# Reformat the data where field name are keys and values are values, each
# dictionary in the list is a new observation

data_formatted = [] # The formatted data
columns = {} # The unique columns and their counts

for obs in data:
    data_formatted.append({})
    for field in obs:
        field_name = field['key'].lower()
        value = field['value']
        data_formatted[-1][field_name] = value
        columns[field_name] = columns.get(field_name, 0) + 1

print('There are {} unique columns'.format(len(columns)))

WRITE_COLUMNS = True
if WRITE_COLUMNS:
    json.dump(columns, open('{}-columns.json'.format(path.stem), 'w'), indent=4)
```

```python
# Create a dataframe from the data use the value key as the value and the key as the column name
df = pd.DataFrame(data_formatted)
print('The dataframe has {} observations'.format(df.shape[0]))
df.head()

if WRITE_DATAFRAME:
    df.to_csv(path.with_suffix('.csv'), index=False)

df.head(20)
```
